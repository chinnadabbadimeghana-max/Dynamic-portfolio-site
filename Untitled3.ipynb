```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

survived = df[df['Survived'] == 1]
age_group_count = survived.groupby('AgeGroup')['Survived'].count()
print(age_group_count)

plt.bar(age_group_count.index, age_group_count.values)
plt.xlabel('Age Group')
plt.ylabel('Number of Survivors')
plt.title('Survivors: Children vs Adults')
plt.show()
```

```python
import matplotlib.pyplot as plt

cities = ['Hyderabad', 'Chennai', 'Mumbai']
population = [85, 90, 200]

labels = ['IT', 'HR', 'Sales']
values = [50, 20, 30]

plt.subplot(1, 2, 1)   # 1 row, 2 columns, 1st plot
plt.bar(cities, population)
plt.title('City Population')

plt.subplot(1, 2, 2)   # 1 row, 2 columns, 2nd plot
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Department Distribution')

plt.show()
```
